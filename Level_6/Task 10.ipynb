```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = [1, 2, 3, 4, 5, 6, 7, 8, 9, 20]
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, color="lightblue", flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8})
plt.title("Box Plot with Outlier Detection")
plt.ylabel("Values")
plt.show()
IQR
```

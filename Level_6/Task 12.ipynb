```python
import numpy as np
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate a sample dataset (normally distributed)
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=100)  # Mean=0, SD=1, 100 samples

# Create a QQ plot
plt.figure(figsize=(6, 4))
stats.probplot(data, dist="norm", plot=plt)
plt.title("QQ Plot for Normality Check")
plt.grid()

# Show the plot
plt.show()
```
